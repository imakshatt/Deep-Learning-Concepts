```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
df.head()
```

```python
x = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked', 'Survived', 'Age', 'Fare'], axis='columns')
```

```python
x.head()
```

```python
y = df.Survived
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = preprocessing.LabelEncoder()
x['Sex'] = label_encoder.fit_transform(x['Sex'])
x
```

```python
x.isnull().sum()
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(n_estimators = 1000, learning_rate = 0.001, max_depth = 15, random_state=1)
```

```python
model.fit(x, y)
```

```python
testdata = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
testdata.head()
```

```python
test = testdata.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked', 'Age', 'Fare'], axis='columns')
```

```python
test.head()
```

```python
label_encoder1 = preprocessing.LabelEncoder()
test['Sex'] = label_encoder1.fit_transform(test['Sex'])
test.head()
```

```python
test.isnull().sum()
```

```python
test.isnull().sum()
```

```python
predictions = model.predict(test)
```

```python
output = pd.DataFrame({'PassengerId': testdata['PassengerId'], 'Survived':predictions})
output.to_csv('submission.csv', index=False)
```
